## What is seaborn?

#### Seaborn is a Python data visualization library based on matplotlib. It provides a high-level interface for drawing attractive and informative statistical graphics.

```python
import seaborn as sns
sns.set(style="ticks")

# Load the example dataset for Ancombe's quartet

df = sns.load_dataset("anscombe")
```

#### Let's check the data with graph

```python
sns.lmplot(x='x', y='y', data=df)
```

```python
sns.lmplot(x='x', y='y', data=df, hue='dataset')
```

```python
# Plot each data with col option

sns.lmplot(x='x', y='y', data=df, hue='dataset', col='dataset')
```

#### The shaded area shows the confidence interval for the distribution of data. You can delete it as an option or make it visible except at both extremes.

```python

sns.lmplot(x='x', y='y', data=df, hue='dataset', col='dataset', ci = None)
```

```python

sns.lmplot(x='x', y='y', data=df, hue='dataset', col='dataset', ci = 50)
# If you don't option the ci, default value is 100
```

```python
sns.lmplot(x='x', y='y', data=df, hue='dataset', col='dataset', ci = 50, col_wrap=2)
```

```python
sns.lmplot(x='x', y='y', data=df, hue='dataset', col='dataset', ci = 50, col_wrap=2, scatter_kws={"s":50, "alpha": 0.1})
```

```python
sns.lmplot(x='x', y='y', data=df, hue='dataset', col='dataset', ci = 50, col_wrap=2, scatter_kws={"s":150, "alpha": 0.1})
```

#### It is an option to adjust the transparency of the points displayed in the scatter plot.

#### S: The size of point, alpha : point transparency
